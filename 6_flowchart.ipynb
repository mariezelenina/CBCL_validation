```python
import pyflowchart as pfc
import numpy as np
```

```python
study_name='ABCD'
study_n = 11812
outside_age_range = 0
inside_age_range = study_n - outside_age_range
missing_data_n = 20
study_not_missing_n = inside_age_range - missing_data_n
positive_n = 119
positive_case_definition = "Depression with or\nwithout comorbidities"
negative_n = study_not_missing_n - positive_n
negative_case_definition = "Anxiety/ADHD but no depression"
neg_samp_n = 78
final_n = positive_n + neg_samp_n
rt = pfc.OperationNode(f"{study_name}\nn = {study_n}")
war = pfc.ConditionNode('Within 8-12 years old?')
warn = pfc.OperationNode(f"Outside Age Range\nn = {outside_age_range}")
wary = pfc.OperationNode(f"n = {inside_age_range}")
md = pfc.ConditionNode('Missing Data?')
mdn = pfc.OperationNode(f"Missing Data\nn={missing_data_n}")
wdn = pfc.OperationNode(f"n={study_not_missing_n}")
pcd = pfc.ConditionNode(f"{positive_case_definition}")
pcn = pfc.OperationNode(f"n = {positive_n}")
#ncd = pfc.ConditionNode(f"{negative_case_definition}")
ncn = pfc.OperationNode(f"{negative_case_definition}\nn = {negative_n}")
rsn = pfc.OperationNode(f"Randomly sample from negative\n so that positive is 10% of sample\n (or all if not enough data)\nn = {neg_samp_n}")
final = pfc.OperationNode(f"Final Sample\nn = {final_n}")

if outside_age_range == 0:
    # if abcd and no age range exclusion
    rt.connect(md)
    md.connect_yes(mdn, direction='right')
    md.connect_no(wdn, direction='bottom')
    wdn.connect(pcd)
    pcd.connect_yes(pcn)
    pcd.connect_no(ncn)
    ncn.connect(rsn)
    pcn.connect(final)
    rsn.connect(final)
    fc = pfc.Flowchart(rt)
    print(fc.flowchart())
else:
    # other studies
    rt.connect(md)
    md.connect_yes(mdn, direction='right')
    md.connect_no(wdn, direction='bottom')
    wdn.connect(war)
    war.connect_no(warn)
    war.connect_yes(wary)
    wary.connect(pcd)
    pcd.connect_yes(pcn)
    pcd.connect_no(ncn)
    ncn.connect(rsn)
    pcn.connect(final)
    rsn.connect(final)
    fc = pfc.Flowchart(rt)
    print(fc.flowchart())
```
