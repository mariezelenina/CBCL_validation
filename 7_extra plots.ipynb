```python
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dataset = ("ABCD:parent", "ABCD_child", "HBN", "BHRC")
hyp_values = {
    'Sensitivity': (0.945, 0.618, 0.888, 0.900),
    'Specificity': (0.861, 0.478, 0.727, 0.783),
    'Strict Specificity': (0, 0.461, 0, 0.781),   
}

x = np.arange(len(dataset))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0
fig, ax = plt.subplots(layout='constrained')

colors=['darkgray','gray','dimgray','lightgray']

for attribute, measurement in hyp_values.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('AUCROC value')
ax.set_title('AUCROC values by dataset')
ax.set_xticks(x + width, dataset)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 1.2)

plt.show()
```

```python
d3 = {'dataset': ['ABCD:parent', 'ABCD:parent', 'ABCD:parent', 'ABCD:child', 'ABCD:child', 'ABCD:child', 'HBN', 'HBN', 'HBN', 'BHRC', 'BHRC', 'BHRC'],
     'hypothesis': ['Sensitivity', 'Specificity', 'Strict Specificity', 'Sensitivity', 'Specificity', 'Strict Specificity', 'Sensitivity', 'Specificity', 'Strict Specificity', 'Sensitivity', 'Specificity', 'Strict Specificity'],
     'AUCROC': [0.953, 0.874, 0,  0.619, 0.478, 0.459, 0.861, 0.712, 0,  0.896, 0.802, 0.781],
     'lci':    [0.927, 0.828, 0,  0.565, 0.398, 0.372, 0.804, 0.648, 0,  0.855, 0.736, 0.697],
     'uci':    [0.974, 0.917, 0,  0.672, 0.558, 0.550, 0.915, 0.774, 0,  0.930, 0.860, 0.855],
     'thresh': [0.8, 0.8,  np.nan,  0.8,   0.8,   0.8, 0.8, 0.8, np.nan,  0.8,   0.8,  0.8]  
     }
df3 = pd.DataFrame(data=d3)
df3['los'] = df3.AUCROC - df3.lci
df3['uos'] = df3.uci - df3.AUCROC

title_lut = {
    'ABCD:parent': 'ABCD: K-SADS Parent',
    'ABCD:child': 'ABCD: K-SADS Child',
    'HBN': 'HBN: K-SADS Clinician',
    'BHRC': 'BHRC: DAWBA Clinician'
}
```

```python
#plotting:
colors=['darkgray','gray','dimgray','lightgray']
g = sns.catplot(x="hypothesis", y="AUCROC",
                 hue = "hypothesis", col="dataset",
                 data=df3, kind="bar",
                 height=4, aspect=.7,
                 palette='Greys');
g.set_xticklabels(rotation=30)


# get some color bar colors so they stand out
foo = sns.palettes.color_palette('Greys', 2)
ecolors =[foo[1], foo[0], foo[0]] 

for ax, (dsname, title) in zip(g.axes[0], title_lut.items()):
    ax.set_title(title)
    ax.set_xlabel('Hypothesis')
    ax.errorbar(
        x=[0,1,2],
        y=df3.loc[df3.dataset == dsname, 'AUCROC'].values,
        yerr=df3.loc[df3.dataset == dsname, ['los', 'uos']].values.T,
        ecolor=ecolors,
        elinewidth=2,
        linestyle='none',
    )
    thrs = df3.loc[df3.dataset == dsname, 'thresh'].values
    for thr, center, color in zip(thrs, [0,1,2], ecolors):
        ax.hlines(thr, center-0.5, center+0.5, colors=color, ls='--')
        
```
